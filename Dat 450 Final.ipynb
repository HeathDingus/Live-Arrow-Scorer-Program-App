```python
# 1. Load the model ONCE before the loop starts
model = YOLO(r"C:\Users\Heath Dingus\runs\detect\train7\weights\best.pt")

cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 2. Run the model on the 'frame' we just captured
    # We use stream=True for better live performance
    results = model(frame, stream=True)

    # 3. Process results and get the annotated frame
    for r in results:
        # This creates a frame with the YOLO boxes already drawn
        annotated_frame = r.plot()

    # 4. Add your custom UI text on top of the annotated frame
    cv2.putText(annotated_frame, "LaneAssist v0.1.0", (15, 100), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)

    # 5. Show the final combined image
    cv2.imshow('LaneAssist Live Feed', annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
