```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from ultralytics import YOLO
from cv2_enumerate_cameras import enumerate_cameras

model = YOLO(r"C:\Users\Heath Dingus\runs\detect\train-37\weights\best.pt")

s = set()
u = []
for x in enumerate_cameras():
    if x.name not in s:
        print(f'Name: {x.name}\n Index: {x.index}') 
        s.add(x.name)
        u.append(x.index)
c = int(input('Select a camera index: '))

cap = cv2.VideoCapture(c)

while cap.isOpened():
    ret, frame = cap.read()
    results = model(frame, stream = True, imgsz = [1080, 1920])
    for x in results:
        x.plot()
    cv2.putText(frame, "LaneAssist v0.1.0", (15, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
    cv2.imshow('LaneAssist', x.plot())
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
