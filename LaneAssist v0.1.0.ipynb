```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from ultralytics import YOLO
import glob
from IPython.display import Image, display

model = YOLO(r"C:\Users\Heath Dingus\runs\detect\train9\weights\best.pt")
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    results = model(frame, stream = True)
    for x in results:
        x.plot()
    cv2.putText(frame, "LaneAssist v0.1.0", (15, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2, cv2.LINE_AA)
    cv2.imshow('LaneAssist', x.plot())

    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
